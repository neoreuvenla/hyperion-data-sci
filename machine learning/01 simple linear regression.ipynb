#### Task 15 - Simple Linear Regression
Performs and plots linear regression

```python
# required imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# global variables for graph colours
data_color = "#006DAF" # blue
fit_color = "#FF6900"  # orange

# global variables for graph size
graph_width = 10
graph_height = 5
```

```python
# load the dataset into a dataframe and inspect
data = pd.read_csv('insurance.csv')
data.head()
```

```python
# scatter plot to visualise age vs charges
plt.figure(figsize=(graph_width,graph_height))
plt.scatter(data['age'], data['charges'], color=data_color)

# titles and labels
plt.xlabel('Age') 
plt.ylabel('Charges')
plt.title('Age vs Charges', weight='bold');
```

```python
# reshape data for sklearn
X = data['age'].values.reshape(-1,1)
y = data['charges'].values.reshape(-1,1)
```

```python
# create a linear regression model
lr_model = LinearRegression()

# train the model on the data 
lr_model.fit(X, y)

# create predictions for x values
y_pred = lr_model.predict(X)
```

```python
# overlay regression line on previous graph
plt.figure(figsize=(graph_width,graph_height))
plt.scatter(data['age'], data['charges'], color=data_color)
plt.plot(X, y_pred, color=fit_color)

# titles and labels
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs Charges', weight='bold');
```
